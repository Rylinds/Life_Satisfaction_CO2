# Comparing Life Satisfaction vs. CO2 Emission Data

```python
# merge datasets
co2_df = pd.read_csv('co2-emissions-per-capita.csv')
```

```python
combined_df = satisfaction_df.merge(
    co2_df,
    left_on=['Country name', 'Year'],
    right_on=['Entity', 'Year'],
    how='inner'
)

# drop duplicate country col
combined_df = combined_df.drop(columns=['Entity'])

combined_df.shape
```

## Best Fit Models

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
x = combined_df[['CO₂ emissions per capita']].values
y = combined_df['Life evaluation (3-year average)'].values
```

```python
# linear fit
lin_model = LinearRegression().fit(x, y)
y_lin = lin_model.predict(x)

r2_linear = r2_score(y, y_lin)
r2_linear
```

```python
# quadratic fit
poly2 = PolynomialFeatures(degree=2)
x_quad = poly2.fit_transform(x)

quad_model = LinearRegression().fit(x_quad, y)
y_quad = quad_model.predict(x_quad)

r2_quadratic = r2_score(y, y_quad)
r2_quadratic
```

```python
# cubic fit
poly3 = PolynomialFeatures(degree=3)
x_cubic = poly3.fit_transform(x)

cubic_model = LinearRegression().fit(x_cubic, y)
y_cubic = cubic_model.predict(x_cubic)

r2_cubic = r2_score(y, y_cubic)
r2_cubic
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y)

# plot best model
plt.plot(np.sort(x, axis=0),
         cubic_model.predict(poly3.transform(np.sort(x, axis=0))))

plt.xlabel("CO2 emissions")
plt.ylabel("Life satisfaction")
plt.title("Best Fit Relationship")
plt.show()
```

## Multiple Linear Regression

```python
features = [
    'Explained by: Log GDP per capita',
    'Explained by: Social support',
    'Explained by: Healthy life expectancy',
    'Explained by: Freedom to make life choices',
    'Explained by: Generosity',
    'Explained by: Perceptions of corruption',
    'CO₂ emissions per capita'
]
```

```python
combined_df[features].isna().sum()
model_df = combined_df[features + ['Life evaluation (3-year average)']].dropna()
```

```python
X = model_df[features].values
y = model_df['Life evaluation (3-year average)'].values
```

```python
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)

r2_score(y, y_pred)
```

```python
pd.Series(model.coef_, index=features)
```
